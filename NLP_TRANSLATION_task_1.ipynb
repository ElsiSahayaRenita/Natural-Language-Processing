```python
import pandas as pd
import nltk
import re

nltk.download('stopwords')
nltk.download("punkt")
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
from deep_translator import GoogleTranslator
```

```python
df = pd.read_csv("English.csv")
df.head()
```

# **Data Cleaning**

```python
#LowerCasing
df['English_refined']= df['English words/sentences'].str.lower()
```

```python
#Removing Special Characters
df['English_refined']= df['English_refined'].str.replace("[^a-zA-Z0-9]", " ") 
df
```

```python
# Duplicates
df["English_refined"] = df["English_refined"].drop_duplicates() 
df.shape
```

```python
df.isnull().sum()
```

```python
#Dropping Null values
df = df.dropna()
```

```python
#New DataFrame taking only last 60 words
dp = df['English_refined'].tail(60)
dp=pd.DataFrame(dp)
```

```python
#Translation Spanish
def translation_spanish(text):
  return GoogleTranslator('auto','es').translate(text)


#Translation French
def translation_french(text):
  return GoogleTranslator('auto','fr').translate(text)

#Translation German
def translation_german(text):
  return GoogleTranslator('auto','de').translate(text)

```

```python
dp["Spanish"] = dp["English_refined"].apply(translation_spanish) #spanish
dp["French"] = dp["English_refined"].apply(translation_french) #french
dp["German"] = dp["English_refined"].apply(translation_german) #german
```
